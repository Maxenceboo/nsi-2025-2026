## üñº 0Ô∏è‚É£ C‚Äôest quoi un arbre binaire ?

Un **arbre binaire** :
- soit **vide** (rien),
- soit un **noeud** avec :
  - une **valeur**,
  - un **fils gauche**,
  - un **fils droit**.

Sch√©ma classique :

```text
        (10)
       /    \
     (5)    (15)
    /  \      \
  (2)  (7)    (20)
```

Chaque fils est **lui-m√™me un arbre binaire** (ou vide) ‚Üí r√©cursif par nature.
