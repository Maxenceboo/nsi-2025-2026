## 1Ô∏è‚É£ C'est quoi une liste ?

Une **liste**, c'est juste un truc qui contient plusieurs valeurs, dans un **ordre pr√©cis**.

Exemples :
- ta **playlist Spotify**
- la **liste des toppings** sur ta pizza
- les **notes** que t'as en NSI (courage)

En Python, √ßa donne par exemple :

```python
playlist = ["Feu rouge", "Ego", "Lose Yourself"]
toppings = ["mozza", "pepperoni", "champignons", "olives"]
notes_nsi = [12, 15, 9, 18]
```

Une liste est entour√©e de `[` et `]`, et les √©l√©ments sont s√©par√©s par des virgules.


````python
# Quelques listes d'exemple
playlist = ["Feu rouge", "Ego", "Lose Yourself"]
toppings = ["mozza", "pepperoni", "champignons", "olives"]
notes_nsi = [12, 15, 9, 18]

print("Playlist :", playlist)
print("Toppings :", toppings)
print("Notes de NSI :", notes_nsi)
````

### Acc√©der aux √©l√©ments d'une liste

Les √©l√©ments sont num√©rot√©s √† partir de **0** :

- `playlist[0]` ‚Üí premier morceau
- `playlist[1]` ‚Üí deuxi√®me morceau
- `playlist[-1]` ‚Üí **dernier** morceau (indice n√©gatif)


````python
print("Premier son de la playlist :", playlist[0])
print("Deuxi√®me son :", playlist[1])
print("Dernier son :", playlist[-1])
````
