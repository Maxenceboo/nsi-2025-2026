## 3ï¸âƒ£ File SANS LISTE, SANS DEQUE : objets chaÃ®nÃ©s

ðŸ‘‰ Ici on fait une vraie **queue** comme en C / Java :
- chaque Ã©lÃ©ment = un **Maillon** qui contient : `valeur` + `suivant`
- la file stocke **tÃªte** et **queue**

Ainsi :
- `enfiler` ajoute APRÃˆS le dernier maillon â†’ O(1)
- `defiler` enlÃ¨ve en tÃªte â†’ O(1)

Câ€™est LA vraie structure FIFO.

### Classe `Maillon`

```python
class Maillon:
    def __init__(self, valeur, suivant=None):
        self.valeur = valeur
        self.suivant = suivant
    def __repr__(self): return f"Maillon({self.valeur})"
```

### Classe `FileObjets` â€“ vraie file chaÃ®nÃ©e

```python
class FileObjets:
    def __init__(self):
        self.tete = None   # premier Ã©lÃ©ment
        self.queue = None  # dernier Ã©lÃ©ment
        self.taille = 0

    def est_vide(self):
        return self.tete is None

    def enfiler(self, valeur):
        nouveau = Maillon(valeur)
        if self.est_vide():
            self.tete = self.queue = nouveau
        else:
            self.queue.suivant = nouveau
            self.queue = nouveau
        self.taille += 1

    def defiler(self):
        if self.est_vide():
            return None
        valeur = self.tete.valeur
        self.tete = self.tete.suivant
        if self.tete is None:
            self.queue = None
        self.taille -= 1
        return valeur

    def tete_valeur(self):
        if self.est_vide(): return None
        return self.tete.valeur

    def __len__(self): return self.taille

    def __repr__(self):
        cur = self.tete
        vals = []
        while cur:
            vals.append(repr(cur.valeur))
            cur = cur.suivant
        return "FileObjets(" + " -> ".join(vals) + ")"
```

### ðŸ§ª Test de la file chaÃ®nÃ©e

```python
f = FileObjets()
f.enfiler('Maxence')
f.enfiler('Perrine')
f.enfiler('Jade')
print(f)
print('Servi:', f.defiler())
print(f)
print('TÃªte:', f.tete_valeur())
print('Taille:', len(f))
```
